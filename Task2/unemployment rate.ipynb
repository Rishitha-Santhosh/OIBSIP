```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('Unemployment in India.csv')
data2=pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
```

```python
print(data1.head())
print(data2.head())
```

```python
print(data1.info())
print(data2.info())
```

```python
print(data1.isnull().sum())
print(data2.isnull().sum())
```

```python
print(data1.describe())
print(data2.describe())
```

```python
print(data1.columns)
```

```python

data1.columns = data1.columns.str.strip()

plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data1, marker='o')
plt.title("Unemployment Rate in India Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
print(data2.columns)
```

```python
data2.columns = data2.columns.str.strip()
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data2, marker='o')
plt.title("Unemployment Rate in India Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
covid_period = data2[data2['Date'] >= "2020-03-01"]
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=covid_period)
plt.title("COVID-19 Impact on Unemployment (India till Nov 2020)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
